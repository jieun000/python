Here are some things you should try:
- 차원을 1로 하여 data를 fit하면 'underfitting'.
- 차원을 6으로 하여 data를 fit하면 너무 많이 fit(overfitting) 되어 실제 test 할 때는 잘 못 맞힘

- 데이터 추가:
    - extreme examples can increase overfitting (assuming they are outliers).
    - 너무 많은 training data가 있으면 overfitting이 증가됨
    - 적당한 training data로 overfitting 감소
- `Regression` 과 `Categorical` 를 스위칭하여 2개의 training sample을 적용함.
