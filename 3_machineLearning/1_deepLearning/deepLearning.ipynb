# perceptron(퍼셉트론) :  신경망(딥러닝)의 기원이 되는 알고리즘
다수의 신호를 입력으로 받아 하나의 신호를 출력

### perceptron(퍼셉트론)의 단순한 논리회로1
- <span style="color: red;">1) AND 게이트</span>
- 2) NAND 게이트
- 3) OR 게이트
- 4) XOR게이트
<img src='images/and.jpg' />

```python
import numpy as np

def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else: return 1

for xs in [(0,0), (1,0), (0,1), (1,1)]:
    y = AND(xs[0], xs[1])
    print(str(xs) + "-> " + str(y))
```

### perceptron(퍼셉트론)의 단순한 논리회로2
- 1) AND 게이트
- <span style="color: red;">2) NAND 게이트 = AND를 NOT 취한 것</span>
- 3) OR 게이트
- 4) XOR게이트
<img src='images/nand.jpg' />

```python
import numpy as np

def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if not(tmp <= 0):
        return 0
    else: return 1

for xs in [(0,0), (1,0), (0,1), (1,1)]:
    y = NAND(xs[0], xs[1])
    print(str(xs) + "->" + str(y))
```

### perceptron(퍼셉트론)의 단순한 논리회로3
- 1) AND 게이트
- 2) NAND 게이트
- <span style="color: red;">3) OR 게이트</span>
- 4) XOR게이트
<img src='images/or.jpg' />

```python
import numpy as np

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else: return 1
    
for xs in [(0,0), (1,0), (0,1), (1,1)]:
    y = OR(xs[0], xs[1])
    print(str(xs) + "->" + str(y))
```

### perceptron(퍼셉트론)의 단순한 논리회로4
- 1) AND 게이트
- 2) NAND 게이트
- 3) OR 게이트
- <span style="color: red;">4) XOR게이트</span>
<img src='images/XOR.jpg' />
#### eXclusive OR => 배타적 OR이 서로 다를 때 1
<img src='images/xor2.jpg' />

```python
import numpy as np

def XOR(x1, x2):
    s1 = NAND (x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2) # nand와 or결과 중 and
    return y

for xs in [(0,0), (1,0), (0,1), (1,1)]:
    y = XOR(xs[0], xs[1])
    print(str(xs) + "->" + str(y))
```

## 이번 장에서 배운 내용
- 퍼셉트론은 입출력을 갖춘 알고리즘, 입력을 주면 정해진 규칙에 값을 출력
- 퍼셉트론은 가중치(weight)와 bias를 파라미터(매개변수, theta)로 설정
- 퍼셉트론으로 AND, OR 게이트 등의 논리 회로 표현
- XOR 게이트는 단층 퍼셉트론으로 표현 불가
- 2층 퍼셉트론을 이용하면 XOR 구현 가능
- 단층 퍼셉트론은 직선형 영역만 표현할 수 있고 다층 퍼셉트론은 비선형 영역도 표현
