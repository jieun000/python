#### 1) AND 연산
-  두 가지가 모두 참(1)이면 결과가 참(1)
- 그 외의 경우는 결과가 거짓(0)

```python
# 교사 학습 프로그램
# 입력(X, Y)와 결과(X and Y)의 모든 패턴을 학습시킴.
# 다시 입력(X, Y)의 모든 패턴을 줬을 때 제대로 된 결과(X and Y)를 내는지 평가함.

# 라이브러리 읽어 들이기.
from sklearn.svm import LinearSVC # LinearSVC 알고리즘 사용 위한 패키지
from sklearn.metrics import accuracy_score # 테스트 결과 평가를 위한 패키지

# 학습 전용 데이터(X, Y)와 결과(X and Y) 준비.
learn_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
learn_label = [0, 0, 0, 1]

# 알고리즘 지정(LinearSVC).
clf = LinearSVC()

# 학습 전용 데이터와 결과 학습.
clf.fit(learn_data, learn_label) # 학습메서드: fit(학습데이터배열, 결과데이터배열)

# 테스트 데이터로 예측.
test_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
test_label = clf.predict(test_data) # 예측 결과 반환: predict(테스트 데이터)

# 예측 결과 평가.
print(test_data, "의 예측 결과는~?", test_label)
                # 정답률 반환: accuracy_score(정답, 예측결과)
print("정답률:", accuracy_score(learn_label, test_label))
```

```python
# XOR 연산
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

learn_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
# xor 전용 레이블로 변경
learn_label = [0, 1, 1, 0]

clf = LinearSVC()

clf.fit(learn_data, learn_label)

test_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
test_label = clf.predict(test_data)

print(test_data, "의 예측 결과는~?", test_label)
print("정답률:", accuracy_score(learn_label, test_label))
```

$=>$ 머신러닝 실패시 접근 방법 선택지
- 알고리즘은 유지하고 매개변수 조정.
- 알고리즘 변경. (★)

```python
from sklearn.neighbors import KNeighborsClassifier # 알고리즘 변경
from sklearn.metrics import accuracy_score

learn_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
learn_label = [0, 1, 1, 0] # xor 전용 레이블

# 알고리즘 변경
clf = KNeighborsClassifier(n_neighbors=1) 

clf.fit(learn_data, learn_label)

test_data = [[0, 0],[1, 0],[0, 1],[1, 1]]
test_label = clf.predict(test_data)

print(test_data, "의 예측 결과는~?", test_label)
print("정답률:", accuracy_score(learn_label, test_label))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
from sklearn.utils import all_estimators

iris_data = pd.read_csv("iris.csv", encoding="utf-8")

x = iris_data.loc[:, ["SepalLength","SepalWidth","PetalLength","PetalWidth"]]
y = iris_data.loc[:, "Name"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, shuffle=True)

# classifier 알고리즘 모두 추출.
warnings.filterwarnings('ignore') # 경고 메시지 무시
allAlgorithms = all_estimators(type_filter="classifier")

for(name, algorithm) in allAlgorithms:
    try:
        clf = algorithm()
        clf.fit(x_train, y_train)
        y_pred = clf.predict(x_test)
        print(name, "의 정답률은:", accuracy_score(y_test, y_pred))
    except Exception as e:
        # print(name, "에서 오류 발생:", str(e))
        pass
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import warnings
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

iris_data = pd.read_csv("iris.csv", encoding="utf-8")

x = iris_data.loc[:, ["SepalLength","SepalWidth","PetalLength","PetalWidth"]]
y = iris_data.loc[:, "Name"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, shuffle=True)

warnings.filterwarnings('ignore')
allAlgorithms = all_estimators(type_filter="classifier")

# kfold_cv: K-분할 크로스 밸리데이션 전용 객체, n_splits: 데이터 분할 수
kfold_cv = KFold(n_splits=5, shuffle=True)

for(name, algorithm) in allAlgorithms:
    try:
        clf = algorithm() # clf: 
        # score 메서드를 가진 클래스를 대상으로.
        if hasattr(clf, "score"): # 알고리즘이 score메서드를 가지고 있다면
            # 크로스 밸리데이션
            # (알고리즘 객체, 입력데이터, 레이블데이터, 크로스 밸리데이션 전용 객체)
            scores = cross_val_score(clf, x, y, cv=kfold_cv)
            print(name, "의 정답률:")
            print(scores)
    except:
        pass
```

```python
# 그리드 서치(Grid Search)를 사용하여 서포트 벡터 머신(SVM)의 하이퍼파라미터를 튜닝
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
# from sklearn.model_selection import RandomizedSearchCV # 랜덤 서치

iris_data = pd.read_csv("iris.csv", encoding="utf-8")

x = iris_data.loc[:, ["SepalLength","SepalWidth","PetalLength","PetalWidth"]]
y = iris_data.loc[:, "Name"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, train_size=0.8, shuffle=True)

# 그리드 서치에서 사용할 매개 변수
parameters = [
    {"C":[1, 10, 100, 1000], "kernel": ["linear"]}, # 선형커널
    {"C":[1, 10, 100, 1000], "kernel": ["rbf"], "gamma":[0.001, 0.0001]}, # 방사형 커널
    {"C":[1, 10, 100, 1000], "kernel": ["sigmoid"], "gamma":[0.001, 0.0001]}, # 시그모이드 커널
]
kfold_cv = KFold(n_splits=5, shuffle=True)

# 그리드 서치
clf = GridSearchCV(SVC(), parameters, cv=kfold_cv)ㄴ
clf.fit(x_train, y_train)
print("최적의 매개 변수는:", clf.best_estimator_)

# 최적의 매개변수로 평가.
y_pred = clf.predict(x_test)
print("최종 정답률:",  accuracy_score(y_test, y_pred))
```
